### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    x = np.linspace(0, 10, 100)
    y = x**2

    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]

    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    if len(categories) != len(values):
        raise ValueError("The number of categories and values must match.")
    if len(x_scatter) != len(y_scatter):
        raise ValueError("Scatter plot inputs must have the same length.")
    if len(labels) != len(sizes):
        raise ValueError("Pie chart labels and sizes must have the same length.")

    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    axs[0, 0].plot(x, y, color='b', linewidth=2)
    axs[0, 0].set_title("Line Chart")

    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title("Bar Chart")

    axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
    axs[1, 0].set_title("Scatter Plot")

    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'orange'], startangle=140, wedgeprops={'edgecolor': 'black'})
    axs[1, 1].set_title("Pie Chart")

    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
